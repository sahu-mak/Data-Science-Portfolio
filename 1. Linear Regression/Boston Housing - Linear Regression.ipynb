```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.DESCR)
boston_df = pd.DataFrame(boston.data)
```

```python
for index, feature_name in enumerate(boston.feature_names):
    plt.figure(figsize=(5, 4))
    plt.scatter(boston.data[:, index], boston.target)
    plt.ylabel('Price', size=12)
    plt.xlabel(feature_name, size=12)
    plt.title(feature_name + ' vs. Price Scatterplot')
    plt.show()
```

```python
corr_matrix = boston_df.corr()
sns.heatmap(corr_matrix,cmap = 'coolwarm')
```

```python
X = boston_df
y = pd.DataFrame(boston.target)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
boston_model = LinearRegression()
```

```python
boston_model.fit(X_train, y_train)
```

```python
#coeff_boston = pd.DataFrame(boston_model.coef_, X.columns, columns = ['Correlation Coefficients'])
#print(coeff_boston)
```

```python
predictions = boston_model.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
#sns.regplot(x = y_test, y = predictions, color = 'red')
```

```python
sns.distplot((y_test-predictions),bins=50);
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
